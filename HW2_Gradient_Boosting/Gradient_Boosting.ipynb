```python
#Преобразуйте все номинальные признаки в несколько признаков с бинарным значение (0.4)
from sklearn.preprocessing import LabelEncoder
data_=data
nominals=['Mjob','Fjob','reason','guardian']
binaryies=['school','sex','address','famsize','Pstatus','schoolsup','famsup','paid','activities','nursery','higher','romantic','internet']
nominal_to_binary_data=pd.get_dummies(data_[nominals])
data_=data_.drop(nominals,axis=1)
data_=pd.concat((data_,nominal_to_binary_data),axis=1)
for b in binaryies:
    data_[b]=LabelEncoder().fit_transform(data_[b]) #Преобразуем значения остальных бинарных признаков к 0 или 1
data_.head(10)
```

```python
#Используя объект SelectKBest с параметром f_regression выберите два признака, которые наиболее сильно связаны с признаком G3 (0.5)
from sklearn.feature_selection import SelectKBest, f_regression
selectKbest_object=SelectKBest(f_regression,k=2).fit(data_.iloc[:, data_.columns != 'G3'],data_['G3'])
print("Наиболее сильно связаны c G3:",data_.columns[selectKbest_object.get_support(indices=True)])
```

Выберите один признак, который наиболее влияет на признак G3 и используя кросс-валидацию (cross-validation) выполните следующее:

- Обучите все полученных объекты на обучающей выборке (0.1)

```python
#На обучающей выборке обучите объект RidgeСV для различных alpha (0.2)
#Выберите alpha при котором RidgeCV лучше всего предсказывает значения (0.1)
#Подсчитайте метрику mean absolute error, mean squared error, median absolute error,r2 score(0.2)
#Max (0.5)
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score
import warnings
warnings.simplefilter('ignore')

alphas = np.linspace(-5, 5, 250)
x = data['G2'].values.reshape(-1, 1)
y = data['G3']
cv=StratifiedKFold(n_splits=2,random_state=1488,shuffle=True)
best_alpha=RidgeCV(alphas=alphas).fit(x,y).alpha_
MAE=RidgeCV(alphas=alphas, cv=cv, scoring=make_scorer(mean_absolute_error)).fit(x, y).score(x,y)
MSE=RidgeCV(alphas=alphas, cv=cv, scoring=make_scorer(mean_squared_error)).fit(x, y).score(x,y)
MedianAE=RidgeCV(alphas=alphas, cv=cv, scoring=make_scorer(median_absolute_error)).fit(x, y).score(x,y)
R2=RidgeCV(alphas=alphas, cv=cv, scoring=make_scorer(r2_score)).fit(x, y).score(x,y)
print("Best Alpha=",best_alpha,"\nMean Absolute=",MAE,"\nMean Squared=",MSE,"\nMedian Absolute",MedianAE,"\nR2 Score=",R2)
```

```python
#Создайте объекты SGDRegressor, используя все возможных варианты аргумента penalty (0.2)
#Обучите все полученных объекты на обучающей выборке (0.1)
#Подсчитайте метрику r2 score для всех объектов (0.2)
#Max (0.5)
from sklearn.linear_model import SGDRegressor
penalties = {'penalty':['none', 'l2', 'l1', 'elasticnet']}
grid_for_sgd=GridSearchCV(cv=cv,estimator=SGDRegressor(),param_grid=penalties,scoring=make_scorer(r2_score))
grid_for_sgd.fit(x,y)
for r2, params in zip(grid_for_sgd.cv_results_['mean_test_score'],  grid_for_sgd.cv_results_['params']):
    print(params,r2,"\n")
```

```python
#Создайте объекты с параметрами max_depth = 1,3, 10 GradientBoostingRegressor и для каждого объекта изменяйте количество n_estimators от 100 до 1000 с шагом 100 (0.2)
#Подсчитайте метрику r2 score для всех объектов (0.1)
#Визуализируйте зависимость R2 oт n_estimators на графике (0.2)
#Max (0.5)
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
max_depth = [1, 3, 10]
n_estimators = range(100, 1000, 100)
for d in max_depth:
    r_scores=[]
    for n_est in n_estimators:
        r_scores.append(GradientBoostingRegressor(n_estimators=n_est,max_depth=d).fit(x,y).score(x,y))
    print("R2 Scores="'{}'.format(r_scores))
    plt.title("max_depth %d" %d) 
    plt.xlabel('n_estimators')
    plt.ylabel('r2_score')
    plt.plot(n_estimators, r_scores, color = 'yellow') 
    plt.grid(True) 
    plt.show()
    print("_________________________________________________________________\n")
```

```python
#Используя все доступные регрессоры в библиотеке sklearn, найдите самый лучший регрессор. (2.5)
from sklearn.linear_model import Lars
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import PassiveAggressiveRegressor
from sklearn.linear_model import ARDRegression

print('LinearRegression',GridSearchCV(cv=cv,estimator=LinearRegression(),param_grid={'normalize': [False, True] }
                                 ,scoring=make_scorer(r2_score)).fit(x,y).score(x,y),'\n')
    
print('LogisticRegression',GridSearchCV(cv=cv,estimator=LogisticRegression(),param_grid= {'penalty': ['l1', 'l2'],
          'C':np.logspace(-4,1,6) },scoring=make_scorer(r2_score)).fit(x,y).score(x,y),'\n')
        
print('Lars:',GridSearchCV(cv=cv,estimator=Lars(),param_grid={'eps': [2e-16,2e-15, 2e-14,2e-13, 2e-12] }
                                 ,scoring=make_scorer(r2_score)).fit(x,y).score(x,y),'\n')
            
print('PassiveAggressiveRegressor',GridSearchCV(cv=cv,estimator=PassiveAggressiveRegressor(),param_grid= {'C':np.logspace(-4,1,6) }
                                 ,scoring=make_scorer(r2_score)).fit(x,y).score(x,y),'\n')
                
print('ARDRegression',GridSearchCV(cv=cv,estimator=ARDRegression(),param_grid=  {'alpha_1': [1e-2,1e-1, 1, 2],
                                                                               'alpha_2': [1e-2,1e-1, 1, 2], 
                                                                               'lambda_1': [1e-2,1e-1, 1, 2], 
                                                                               'lambda_2': [1e-2,1e-1, 1, 2] } ,
                                   scoring=make_scorer(r2_score)).fit(x,y).score(x,y),'\n')
    
```

```python
#В статье описано использование Random Forest для предсказания важности факторов, влияющих на потребление алкоголя. Повторите эксперимент с использование объекта RandomForest
#и найдите наилучшие параметры. Укажите accuracy для наилучших параметров (0.5)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
data2=data_
y_new=[]
for i in data2.index:
    if((data['Walc'][i]*5+data['Dalc'][i]*2)/7 < 3):
        y_new.append(0)
    elif((data['Walc'][i]*5+data['Dalc'][i]*2)/7 >0):
        y_new.append(1)
del data2['Walc'], data2['Dalc'],data2['G1'],data2['G2'],data2['G3']
x_new=data2

paramsrf = {'n_estimators': range(10, 100, 10),'max_depth': [1,3,5,7],'min_samples_leaf': [2,3,4],'min_samples_split': [2,3,4]}
clf_random_forest=GridSearchCV(cv=cv,estimator=RandomForestClassifier(random_state=1488),scoring='accuracy',param_grid=paramsrf)
clf_random_forest.fit(x_new,y_new)
rf_accuracy = accuracy_score(y_new, clf_random_forest.predict(x_new))
top_rf_params = clf_random_forest.best_params_
print("Accuracy=",rf_accuracy,"\nBest Params:",top_rf_params)
```

```python
#Обучите объект GradientBoostingClassifier и найдите для него наилучшие параметры. Укажите accuracy для наилучших параметров (0.5)
from sklearn.ensemble import GradientBoostingClassifier
paramsgbc = {'loss': ['deviance', 'exponential'], 'n_estimators': range(10, 100, 10),
                  'max_depth': [1,3,5,7],'min_samples_split': [2,3,4],'min_samples_leaf': [2,3,4]}
clf_gradient=GridSearchCV(cv=cv,estimator=GradientBoostingClassifier(random_state=1488),scoring='accuracy',param_grid=paramsgbc)
clf_gradient.fit(x_new,y_new)
gbs_accuracy=accuracy_score(y_new,clf_gradient.predict(x_new))
top_gbs_params=clf_gradient.best_params_
print("Accuracy=",gbs_accuracy,"\nBest Params:",top_gbs_params)
```

```python
#Обучите объект AdaBoostClassifier и найдите для него наилучшие параметры. Укажите accuracy для наилучших параметров (0.5)
from sklearn.ensemble import AdaBoostClassifier
paramsabc={'n_estimators' : range(10,100,10),'algorithm' : ['SAMME', 'SAMME.R'],'learning_rate' : np.linspace(0.5, 4, 8)}
clf_ada=GridSearchCV(cv=cv,estimator=AdaBoostClassifier(),scoring='accuracy',param_grid=paramsabc)
clf_ada.fit(x_new,y_new)
ada_accuracy=accuracy_score(y_new,clf_ada.predict(x_new))
top_ada_params=clf_ada.best_params_
print("Accuracy=",ada_accuracy,"\nBest Params:",top_ada_params)
```
