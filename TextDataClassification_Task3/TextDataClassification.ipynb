```python
#1
import numpy as np
import pandas as pd

data=pd.read_table("data/reviews.tsv")
data
```

```python
x=data.ix[:,1].values
y=data.ix[:,0].values
x
```

```python
#2,3
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
vectorization = TfidfVectorizer(lowercase=True, ngram_range=(1,1))
x_v = vectorization.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x_v,y,test_size=0.2,random_state=123)
```

```python
#4
#Постройте на обучающем подмножестве следующие модели классификации:
#K-ближайших соседей ( n=5n=5 )
#Логистическая регрессия ( C=1C=1 )
#Мультиномиальный наивный Байес ( α=1α=1 )
#Определите качество классификации (по доле правильных классификаций) на тестовом подмножестве
#Определите с помощью timeit время обучения и предсказания (на тестовом подмножестве)
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import timeit

#Trainings
def k_neighbors_training():
    knn=KNeighborsClassifier(n_neighbors=5)
    model=knn.fit(x_train,y_train)
    return model
def logistic_training():
    log=LogisticRegression(penalty="l2" ,fit_intercept=True,max_iter=100,C=1,solver="lbfgs",random_state=12345)
    model=log.fit(x_train,y_train)
    return model
def multinom_training():
    mult=MultinomialNB(alpha=1)
    model=mult.fit(x_train,y_train)
    return model

#Predictions
def k_neighbors_predictions():
    model=k_neighbors_training()
    predictions=model.predict(x_test)
    return predictions
def logistic_predictions():
    model=logistic_training()
    predictions=model.predict(x_test)
    return predictions
def multinom_predictions():
    model=multinom_training()
    predictions=model.predict(x_test)
    return predictions


print("====Time====")
#Timer
print("K neigbors training:", timeit.Timer(k_neighbors_training).timeit(1))
print("Logistic Regression training:", timeit.Timer(logistic_training).timeit(1))
print("Multinomial training:",timeit.Timer(multinom_training).timeit(1))
print("K neigbors predictions:",timeit.Timer(k_neighbors_predictions).timeit(1))
print("Logistic Regression predictions:",timeit.Timer(logistic_predictions).timeit(1))
print("Multinomial predictions:",timeit.Timer(multinom_predictions).timeit(1))

print("\n===Accuracy===")
#Accuracy
kn_accuracy=accuracy_score(y_test,k_neighbors_predictions())
log_accuracy=accuracy_score(y_test,logistic_predictions())
mul_accuracy=accuracy_score(y_test,multinom_predictions())
print("K neigbors accuracy:",kn_accuracy)
print("Logistic Regression accuracy:",log_accuracy)
print("Multinomial accuracy:",mul_accuracy)
```

<p><b>Задача 2.</b> Применение k-folds <b><i>(3 балла)</i></b></p>

<p>Повторите решение задачи 1, но с ипользованием k-folds (k=4) для разделения исходных данных, обучения и тестирования.</p>

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

kfold=KFold(n_splits=4,shuffle=True, random_state=123)
knn=KNeighborsClassifier(n_neighbors=5)
log=LogisticRegression(penalty="l2" ,fit_intercept=True,max_iter=100,C=1,solver="lbfgs",random_state=12345)
mult=MultinomialNB(alpha=1)

def crosval_kn():
    knn_accuracy=cross_val_score(knn,x_v,y,scoring="accuracy",cv=kfold).mean()
    return knn_accuracy
def crosval_logr():
    logr_accuracy=cross_val_score(log,x_v,y,scoring="accuracy",cv=kfold).mean()
    return logr_accuracy
def crosval_mult():
    mult_accuracy=cross_val_score(mult,x_v,y,scoring="accuracy",cv=kfold).mean()
    return mult_accuracy

#Timer and mean accuracy
print("==K neigbors:==\nTime:", timeit.Timer(crosval_kn).timeit(1),",Mean Accuracy:",crosval_kn())
print("\n==Logistic Regression==:\nTime:", timeit.Timer(crosval_logr).timeit(1),",Mean Accuracy:",crosval_logr())
print("\n==Multinomial:==\nTime:",timeit.Timer(crosval_mult).timeit(1),",Mean Accuracy:",crosval_mult())
```
