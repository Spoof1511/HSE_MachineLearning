<p><b>Задача 1.</b> Регрессия + кросс-валидация</p>

<p>Дано множество наблюдений (файл "Reg_A5.csv"), регрессор - линейная регрессия. Найти степень полинома с минимальной ошибкой на проверочном подможестве, определить среднеквадратическую ошибку на тестовом подмножестве? (Степень полинома от 1 до 25)</p>
<p>
Построить графики:<br>
1) зависимости <b>среднеквадратической ошибки</b> от степени полинома для обучающего и проверочного подмножеств<br>
2) зависимости <b>R-квадрата</b> от степени полинома для обучающего и проверочного подмножеств<br>
3) функцию регрессии (наилучший случай) + исходные данные</p>

```python
#0) Подключение библиотек
import pandas as pnd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter('ignore')

from matplotlib.colors import ListedColormap
clrMap = ListedColormap(["blue", "red", "green"])
```

<p>1. Загрузка исходных данных</p>

```python
df = pnd.read_csv("data/Reg_A5.csv")

x = df.X.values.reshape(-1, 1)
y = df.Y.values.reshape(-1, 1)

plt.title("Initial Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, y, "o")
plt.grid(True)
plt.show()
```

<p>2. Формирование обучающего (train), проверочного (val) и тестового (test) подмножеств двумя способами</p>

```python
# Способ 1
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_predict, train_test_split

x_train_init, x_test, y_train_init, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
x_train, x_val, y_train, y_val = train_test_split(x_train_init, y_train_init, test_size=0.25, random_state=0)

# Способ 2
# Воспользоваться KFold с параметрами n=len(x_train_init), n_folds=4, shuffle=True, random_state=0
kfold=KFold(n=len(x_train_init),n_folds=4,shuffle=True,random_state=0)
kfold
```

<p>3. Обучение</p>

```python
#LinearRegression с нормализацией X
from sklearn.linear_model import LinearRegression,LogisticRegression, RidgeCV, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, accuracy_score, mean_squared_error
from numpy.polynomial.polynomial import polyval
from sklearn.pipeline import Pipeline

# Для способов 1 и 2

# Для способа 2
#Для каждой степени полинома ошибка обучения:
#msetr = 1/4*(ErrorTrainFold_1+ErrorTrainFold_2+ErrorTrainFold_3+ErrorTrainFold_4)

#Для каждой степени полинома ошибка на проверочном подмножестве:
#msev = 1/4*(ErrorValFold_1+ErrorValFold_2+ErrorValFold_3+ErrorValFold_4)



#Для 1 способа
mse_train1=[]
mse_val1=[]
r_train1=[]
r_val1=[]
for d in range(1,26): #степень полинома от 1 до 25
       #Обучение
    model1=Pipeline([("pFeatures", PolynomialFeatures(degree=d)), ("lr", LinearRegression(fit_intercept=True,normalize=True))])
    model1=model1.fit(x_train,y_train)
       #Подсчет ошибки обучения и R квадрат
    pred_train=model1.predict(x_train)
    pred_val=model1.predict(x_val)
    mse_train=mean_squared_error(y_train, pred_train)
    mse_val=mean_squared_error(y_val,pred_val)
    r2_train=r2_score(y_train, pred_train)
    r2_val=r2_score(y_val,pred_val)
    mse_train1.append(mse_train)
    mse_val1.append(mse_val)
    r_train1.append(r2_train)
    r_val1.append(r2_val)

#Для 2 способа
mse_train2=[]
mse_val2=[]
r_train2=[]
r_val2=[]
for d in range(1,26):
         #Обучение
    model2=Pipeline([("pFeatures", PolynomialFeatures(degree=d)), ("lr", LinearRegression(fit_intercept=True,normalize=True))])
    mse_train=0
    mse_val=0
    r2_train=0
    r2_val=0
    for index_tr,index_val in kfold:
        x_tr=x_train_init[index_tr].reshape(-1,1)
        x_v=x_train_init[index_val].reshape(-1,1)
        y_tr=y_train_init[index_tr]
        y_v=y_train_init[index_val]
        model2.fit(x_tr,y_tr)
          #Подсчет ошибки обучения и R квадрат
        pred_train=model2.predict(x_tr)
        mse_train=mse_train+1/4*mean_squared_error(y_tr,pred_train)
        r2_train=r2_train+1/4*r2_score(y_tr,pred_train)
        pred_val=model2.predict(x_v)
        mse_val=mse_val+1/4*mean_squared_error(y_v,pred_val)
        r2_val=r2_val+1/4*r2_score(y_v,pred_val)
    mse_train2.append(mse_train)
    mse_val2.append(mse_val)
    r_train2.append(r2_train)
    r_val2.append(r2_val)
    
```

<p>4. Графики зависимости среднеквадратической ошибки от степени полинома для обучающего и проверочного подмножеств</p>

```python
# Для способов 1 и 2

#Ваш код
#1 способ
print("\n          1 способ:\n")
plt.figure(figsize=[13,5])
plt.subplot(1,2,1)
x_plt_tr=range(25)
y_plt_tr=mse_train1
plt.xlabel("degree")
plt.ylabel("MSE")
plt.title("train")
plt.plot(x_plt_tr,y_plt_tr)
plt.grid(True)

plt.subplot(1,2,2)
x_plt_val=range(25)
y_plt_val=mse_val1
plt.title("validation")
plt.xlabel("degree")
plt.ylabel("MSE")
plt.plot(x_plt_val, y_plt_val)
plt.grid(True)
plt.show()


#2 способ
print("\n          2 способ:\n")
plt.figure(figsize=[13,5])
plt.subplot(1,2,1)
x_plt_tr=range(25)
y_plt_tr=mse_train2
plt.xlabel("degree")
plt.ylabel("MSE")
plt.title("train")
plt.plot(x_plt_tr,y_plt_tr)
plt.grid(True)

plt.subplot(1,2,2)
x_plt_val=range(25)
y_plt_val=mse_val2
plt.title("validation")
plt.xlabel("degree")
plt.ylabel("MSE")
plt.plot(x_plt_val, y_plt_val)
plt.grid(True)
plt.show()
```

<p>5. Графики зависимости R-квадрата от степени полинома для обучающего и проверочного подмножеств</p>

```python
# Для способов 1 и 2

#Ваш код
#1 способ
print("\n          1 способ:\n")
plt.figure(figsize=[13,5])
plt.subplot(1,2,1)
x_plt_tr=range(25)
y_plt_tr=r_train1
plt.xlabel("degree")
plt.ylabel("R ^2")
plt.title("train")
plt.plot(x_plt_tr,y_plt_tr)
plt.grid(True)

plt.subplot(1,2,2)
x_plt_val=range(25)
y_plt_val=r_val1
plt.title("validation")
plt.xlabel("degree")
plt.ylabel("R ^2")
plt.plot(x_plt_val, y_plt_val)
plt.grid(True)
plt.show()


#2 способ
print("\n          2 способ:\n")
plt.figure(figsize=[13,5])
plt.subplot(1,2,1)
x_plt_tr=range(25)
y_plt_tr=r_train2
plt.xlabel("degree")
plt.ylabel("R ^2")
plt.title("train")
plt.plot(x_plt_tr,y_plt_tr)
plt.grid(True)

plt.subplot(1,2,2)
x_plt_val=range(25)
y_plt_val=r_val2
plt.title("validation")
plt.xlabel("degree")
plt.ylabel("R ^2")
plt.plot(x_plt_val, y_plt_val)
plt.grid(True)


plt.show()
```

<p>6. Выбор степени полинома + повторное обучение модели на train_init</p>

```python
# Для способов 1 и 2

#Ваш код

#1 способ
best_degree1=0
for m in range(len(mse_val1)):
    if (mse_val1[m]==min(mse_val1)):
        best_degree1=m+1
print("Mean Squared Error(val)(1 способ)=",min(mse_val1))
print("The best degree(1 способ)=",best_degree1)

nmodel1 = Pipeline([("plFeatures", PolynomialFeatures(degree=best_degree1)), ("lr", LinearRegression(fit_intercept=True, normalize=True))])
nmodel1.fit(x_train_init, y_train_init)       
        
#2 способ
best_degree2=0
for m in range(len(mse_val2)):
    if (mse_val2[m]==min(mse_val2)):
        best_degree2=m+1
print("Mean Squared Error(val)(2способ)=",min(mse_val2))
print("The best degree(2 способ)=",best_degree2)

nmodel2 = Pipeline([("plFeatures", PolynomialFeatures(degree=best_degree2)), ("lr", LinearRegression(fit_intercept=True, normalize=True))])
nmodel2.fit(x_train_init, y_train_init)
print("\n")
```

<p>7. Cреднеквадратическая ошибка на тестовом подмножестве</p>

```python
# Для способов 1 и 2

#Ваш код
#1 способ
pred_test_1=nmodel1.predict(x_test)
mse_test_1=mean_squared_error(y_test,pred_test_1)

#2 способ
pred_test_2=nmodel2.predict(x_test)
mse_test_2=mean_squared_error(y_test,pred_test_2)

print("Mean Squared Error(test)(1способ)=",mse_test_1)
print("Mean Squared Error(test)(2способ)=",mse_test_2)
```

<p>8. График функции регрессии (наилучший случай) + исходные данные</p>

```python
# Для способов 1 и 2
import numpy as np
#Ваш код
plt.figure("1",figsize=[14,5])
#1 способ
plt.subplot(1,2,1)
plt.xlabel("X")
plt.ylabel("Y")
xx = np.linspace(0,5,100).reshape(-1, 1)
plt.title("Regression graphic:(1 sposob):")
plt.plot(x,y,"o")
plt.plot(xx, nmodel1.predict(xx),color="yellow", lw=3)
plt.grid(True)

#2 способ
plt.subplot(1,2,2)
plt.xlabel("X")
plt.ylabel("Y")
xx = np.linspace(0,5,100).reshape(-1, 1)
plt.title("Regression graphic:(2 sposob):")
plt.plot(x,y,"o")
plt.plot(xx, nmodel2.predict(xx),color="yellow", lw=3)
plt.grid(True)

plt.show()
```

<p>9. Сравнение результатов для различных способов</p>

```python
# Среднеквадратическая ошибка для метода Kfold получилась меньше и из-за этого результат получился лучше,чем у 1 способа,
# так как Кфолд помимо разделения данных еще и перемешивает их и потом усредняет ошибку и это дает более точный результат.
```

<p><b>Задача 2.</b> Классификация + кросс-валидация</p>

<p>Дано множество наблюдений (файл "Cl_A5_Vx.csv"), классификатор - логистическая регрессия. Найти степень полинома с минимальной ошибкой на проверочном подможестве, определить долю правильных классификаций на тестовом подмножестве?</p>
<p>
Построить графики:<br>
1) зависимости доли правильных классификаций от <b>степени полинома</b> для обучающего и проверочного подмножеств<br>
2) зависимости доли правильных классификаций от <b>количества итераций</b> для обучающего и проверочного подмножеств для наилучшего случая<br>
3) результат классификации для наилучшего случая для обучающего и проверочного подмножеств</p>
4) результат классификации для тестового подмножества</p>

<p>1. Загрузка исходных данных</p>

```python
df = pnd.read_csv("data/Cl_A5_V"+str(variant)+".csv") #variant - вариант задания

plt.title("Initial Data")
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(df.X1, df.X2, c=df.Y, cmap=clrMap)
plt.grid(True)
plt.show()


x1 = [n for n in df.X1]
x2 = [n for n in df.X2]


x=np.column_stack((x1,x2 ))
y = [n for n in df.Y]
```

<p>2. Формирование обучающего (train), проверочного (val) и тестового (test) подмножеств</p>

```python
# Способ 1
x_train_init, x_test, y_train_init, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
x_train, x_val, y_train, y_val =train_test_split(x_train_init, y_train_init, test_size=0.25, random_state=0)

# Способ 2
# Воспользоваться KFold с параметрами n=len(x_train_init), n_folds=4, shuffle=True, random_state=0

#Ваш код
kfold2=KFold(n=len(x_train_init), n_folds=4, shuffle=True, random_state=0)
```

<p>3. Обучение</p>

```python
#Параметры классификатора
#penalty="l2", fit_intercept=True, max_iter=x, C=1e5, solver="lbfgs", random_state=12345

# Для способов 1 и 2

# Для способа 2
#Для каждой степени полинома ошибка обучения:
#msetr = 1/4*(ErrorTrainFold_1+ErrorTrainFold_2+ErrorTrainFold_3+ErrorTrainFold_4)

#Для каждой степени полинома ошибка на проверочном подмножестве:
#msev = 1/4*(ErrorValFold_1+ErrorValFold_2+ErrorValFold_3+ErrorValFold_4)

#Ваш код

#Для 1 способа
mse_train1=[]
mse_val1=[]
acc_train1=[]
acc_val1=[]
for d in range(1,26):
    model_1=Pipeline([("pFeatures", PolynomialFeatures(degree=d)),("lr", LogisticRegression(penalty="l2", fit_intercept=True, max_iter=100, C=1e5, solver="lbfgs", random_state=12345))])
    model_1=model_1.fit(x_train,y_train)
    pred_train=model_1.predict(x_train)
    mse_tr=mean_squared_error(y_train,pred_train)
    acc_tr=accuracy_score(y_train,pred_train)
    pred_val=model_1.predict(x_val)
    mse_val=mean_squared_error(y_val,pred_val)
    acc_val=accuracy_score(y_val,pred_val)
    mse_train1.append(mse_tr)
    mse_val1.append(mse_val)
    acc_train1.append(acc_tr)
    acc_val1.append(acc_val)

#Для 2 способа
mse_train2=[]
mse_val2=[]
acc_val2=[]
acc_train2=[]
for i in range(1,26):   
    model_2 = Pipeline([("plFeatures", PolynomialFeatures(degree=i)), ("lr", LogisticRegression(penalty="l2", fit_intercept=True, max_iter=100, C=1e5, solver="lbfgs", random_state=12345))])  
    msetr=0
    msev=0
    acctr=0
    accval=0
    for train_index, val_index in kfold2: 
        x_tr2 = x_train_init[train_index]
        x_v2 = x_train_init[val_index]  
        y_tr2=[]
        y_v2=[]
        y_pred_train=[]
        y_pred_val=[] 
        for i in range(len(train_index)):
            y_tr2.append(y_train_init[train_index[i]])
        for i in range(len(val_index)):
            y_v2.append(y_train_init[val_index[i]])   
        model_2.fit(x_tr2, y_tr2) 
        y_pred_train=model_2.predict(x_tr2)
        msetr=msetr+1/4*mean_squared_error(y_tr2, y_pred_train)
        acctr=acctr+1/4*model_2.score(x_tr2, y_tr2)
        y_pred_val=model_2.predict(x_v2)
        msev=msev+1/4*mean_squared_error(y_v2, y_pred_val)
        accval=accval+1/4*model_2.score(x_v2, y_v2)
    mse_val2.append(msev)
    mse_train2.append(msetr)
    acc_train2.append(acctr)
    acc_val2.append(accval)
```

<p>4. Графики зависимости доли правильных классификаций от степени полинома для обучающего и проверочного подмножеств</p>

```python
# Для способов 1 и 2

#Ваш код
plt.figure("1",figsize=[13,5])
#1 способ
plt.subplot(1,2,1)
plt.xlabel("degree")
plt.ylabel("accuracy")
plt.title("1 sposob")
plt.plot(range(1,26),acc_train1,lw=2,label="train")
plt.plot(range(1,26),acc_val1,lw=2,label="validation")
plt.legend(loc=1,fontsize=14)
plt.grid(True)

#2 способ
plt.subplot(1,2,2)
plt.xlabel("degree")
plt.ylabel("accuracy")
plt.title("2 sposob")
plt.plot(range(1,26),acc_train2,lw=2,label="train")
plt.plot(range(1,26),acc_val2,lw=2,label="validation")
plt.legend(loc=1,fontsize=14)
plt.grid(True)
```

<p>5. Графики зависимости доли правильных классификаций от количества итераций для обучающего (train) и проверочного подмножеств для наилучшего случая - изменение параметра max_iter</p>

```python

#Ваш код
if (np.max(acc_val1)<np.max(acc_val2)):
    print("Use 1 sposob!")
else:
    print("Use 2 sposob!")
tracc=[]
valacc=[]

for i in range(1,101):
      modelka = Pipeline([("plFeatures", PolynomialFeatures(1)), ("lr", LogisticRegression(max_iter=i, C=1e5, solver="lbfgs", random_state=12345))]) 
      modelka.fit(x_train,y_train)
      tracc.append(accuracy_score(y_train,modelka.predict(x_train)))
      valacc.append(accuracy_score(y_val,modelka.predict(x_val)))
plt.figure("1",figsize=[13,5])
plt.subplot(1,2,1)
plt.title("train")
plt.xlabel("iteration")
plt.ylabel("accuracy")
plt.plot(range(1, 101), tracc,lw=2)
plt.grid(True)

plt.subplot(1,2,2)
plt.title("validation")
plt.xlabel("iteration")
plt.ylabel("accuracy")
plt.plot(range(1, 101), valacc, lw=2)
plt.grid(True)
```

<p><b>Задача 3.</b> Регрессия + регуляризация</p>

<p>1. Загрузка исходных данных</p>

```python
df = pnd.read_csv("data/Reg_A5.csv")

x = df.X.values
y = df.Y.values

plt.title("Initial Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, y, "o")
plt.grid(True)
plt.show()
```

<p>2. Обучение с кросс-валидацией (см. задачу 1 способ 2) и с использованием Ridge при различных значениях коэф. регуляризации</p>

```python
# Регрессор - Ridge
# Параметры: a fit_intercept=True, normalize=True, random_state=12345
# Здесь и далее СТЕПЕНЬ ПОЛИНОМА - 16!!!

#Коэф. регуляризации
alphas = np.append([0.0], np.logspace(-8, 1, 20, base=10))

#Для каждого значения коэф. регуляризации:
#msetr = 1/4*(ErrorTrainFold_1+ErrorTrainFold_2+ErrorTrainFold_3+ErrorTrainFold_4)

#Для каждой степени полинома ошибка на проверочном подмножестве:
#msev = 1/4*(ErrorValFold_1+ErrorValFold_2+ErrorValFold_3+ErrorValFold_4)

#Ваш код
x_train_init, x_test, y_train_init, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
kfoldr = KFold(n=len(x_train_init), n_folds=4, shuffle=True, random_state=0)

mseR_train=[]
mseR_val=[]
mseR_test=[]
rr2_train=[]
rr2_val=[]
for i in alphas:
    model_ridge = Pipeline([("plFeatures", PolynomialFeatures(degree=16)), ("lr", Ridge(alpha=i, fit_intercept=True, normalize=True, random_state=12345))])
    mseRtr=0
    mseRtest=0
    mseRval=0
    r2_val=0
    r2_train=0
    for train_index, val_index in kfoldr: 
        y_train=[]
        y_val=[]
        x_train = x_train_init[train_index].reshape(-1, 1)
        x_val = x_train_init[val_index].reshape(-1, 1)
        for i in range(len(train_index)):
            y_train.append(y_train_init[train_index[i]])
        for i in range(len(val_index)):
            y_val.append(y_train_init[val_index[i]])
            
        model_ridge.fit(x_train, y_train)    
        pred_train=model_ridge.predict(x_train)
        mseRtr=mseRtr+1/4*mean_squared_error(y_train, pred_train)
        r2_train=r2_train+1/4*r2_score(y_train, pred_train)
        pred_val=model_ridge.predict(x_val)
        mseRval=mseRval+1/4*mean_squared_error(y_val, pred_val)
        r2_val=r2_val+1/4*r2_score(y_val, pred_val)
        x_test=x_test.reshape(-1,1)
        pred_test2=model_ridge.predict(x_test)
        mseRtest=mseRtest+1/4*mean_squared_error(y_test, pred_test2)
        
    mseR_val.append(mseRval)
    mseR_train.append(mseRtr)
    mseR_test.append(mseRtest)
    rr2_train.append(r2_train)
    rr2_val.append(r2_val)
```

<p>3. Графики зависимости среднеквадратической ошибки от степени полинома для обучающего и проверочного подмножеств</p>

```python
plt.figure("1", figsize=[13,5])

plt.subplot(1,2,1)
plt.title("mse train")
plt.xlabel("alpha")
plt.ylabel("mse")
plt.plot(range(0, 21), mseR_train, lw=2)
plt.grid(True)

plt.subplot(1,2,2)
plt.title("mse validation")
plt.xlabel("alpha")
plt.ylabel("mse")
plt.plot(range(0, 21), mseR_val,lw=2)
plt.grid(True)
```

<p>4. Графики зависимости R-квадрата от степени полинома для обучающего и проверочного подмножеств</p>

```python
plt.figure("1", figsize=[13,5])

plt.subplot(1,2,1)
plt.title("mse train")
plt.xlabel("alpha")
plt.ylabel("R2")
plt.plot(range(0, 21), rr2_train,lw=2)
plt.grid(True)

plt.subplot(1,2,2)
plt.title("mse validation")
plt.xlabel("alpha")
plt.ylabel("R2")
plt.plot(range(0, 21), rr2_val, lw=2)
plt.grid(True)
```

```python
print(' коэф. рег. с минимальной ошибкой на val {}. alpha {}'.format(np.min(mseR_val), alphas[np.argmin(mseR_val)]))
```

<p>8. Обучение на train_init с использованием Ridge и параметром alpha=(значение из пункта 2)</p>

```python
bridge_model = Pipeline([("plFeatures", PolynomialFeatures(degree=16)), ("lr", Ridge(alpha=alphas[np.argmin(mseR_val)], fit_intercept=True, normalize=True, random_state=12345))])
x_train_init = x_train_init.reshape(-1, 1)
bridge_model.fit(x_train_init, y_train_init)
print()
```
