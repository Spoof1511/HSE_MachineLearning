<p><b>Задача 2.</b> Анализ рынка акций</p>

<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
import glob
import pandas as pnd
import os
import numpy as np
path=r'data\\stock'
names=os.listdir(path)
filenames=glob.glob(path + "/*.csv")
df_act=pnd.DataFrame()

i=0

for file in filenames:
    temp=pnd.read_csv(file,index_col='Date')
    df_act[i]=temp['Close']
    i+=1
i=0
for name in names:
    df_act.rename(index=str, columns={i:name[:-4]},inplace=True)
    i+=1
df_act      

    
```

<p>2. Рассчитайте корреляционную матрицу для всех акций</p>

```python
corrm=df_act.corr()
corrm
```

<p>3. Отобразите корреляционную матрицив в виде диаграммы. Примерный вид графика приведен ниже</p>
<img src="img/fig_corr_matrix.png" width="400px">

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

labls=df_act.columns.values.tolist() 
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corrm, vmin=-1, vmax=1,cmap = plt.cm.ocean_r)
fig.colorbar(cax)
ax.set_xticklabels(['']+labls,rotation='vertical',size=8)
ax.set_yticklabels(['']+labls,size=8)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
plt.show()
```

<p>4. В соответствии с Вашим вариантом определите:</p> 
<ul>
    <li>акцию с максимальной корреляцией (max)</li>
    <li>акцию с минимальной корреляцией (min) </li>
    <li>акцию, которая больше всего соответствует отсутствию какой-либо корреляции (none)</li>

<p><b>Вариант 1.</b> Microsoft (MSFT)</p>
<p><b>Вариант 2.</b> Facebook (FB)</p>
<p><b>Вариант 3.</b> Google (GOOGL)</p>
<p><b>Вариант 4.</b> Alibaba (BABA)</p>

```python
corr_temp=corrm.copy()
del corr_temp['MSFT']['MSFT']
msft_df=corr_temp['MSFT']
maximum=min(range(len(msft_df)),key=lambda i :abs(msft_df[i]-1))
minimum=min(range(len(msft_df)),key=lambda i:abs(msft_df[i]+1))
closest_to_zero=min(range(len(msft_df)), key=lambda i: abs(msft_df[i]-0))
print("Акция с none корреляцией:",msft_df.index[closest_to_zero])
print("Акция с min корреляцией:",msft_df.index[maximum])
print("Акция с max корреляцией:",msft_df.index[minimum])
```

<p>5. Постройте диаграммы разброса (Ваша компания - Компания с min), (Ваша компания - Компания с max), (Ваша компания - Компания с none)</p>

```python
plt.scatter(corrm['MSFT'], corrm[[msft_df.index[maximum]]], c = ['black','red']) 
plt.title('Microsoft and Max Company') 
plt.show()
plt.scatter(corrm['MSFT'],corrm[[msft_df.index[minimum]]],c=['black','orange'])
plt.title('Microsoft and Min Company')
plt.show()
plt.scatter(corrm['MSFT'],corrm[[msft_df.index[closest_to_zero]]],c=['black','yellow'])
plt.title('Microsoft and None Company')
plt.show()
```
